```python
dataPath = '../Data/Raw/final_books_data_2011_2024.csv'
file = open(dataPath, 'r')
cols = {}

firstLine = file.readline().rstrip().split(',')
print(firstLine)

for colName in firstLine:
    cols[colName] = []

for line in file:
    lineVals = line.rstrip().split(',')

    l = len(lineVals)

    curIdx = 0
    for i in range(6):
        value = lineVals[curIdx]
        if '"' in value:
            curIdx = i + 1

            while not ('"' in lineVals[curIdx]):
                value += ', ' + lineVals[curIdx]
                curIdx += 1

            value += ', ' + lineVals[curIdx]
            value = value.replace('"', '').strip()

        cols[firstLine[i]].append(value)
        curIdx += 1

    genreEnd = l - 17
    genreCount = genreEnd - curIdx
    genres = []

    for i in range(genreCount):
        genre = lineVals[curIdx]
        charToReplace = ['[', ']', '\'', '"']
        
        for char in charToReplace:
            genre = genre.replace(char, '').strip()

        genres.append(genre)
        curIdx += 1

    cols[firstLine[6]].append(genres)

    for i in range(-17, 0, 1):
        cols[firstLine[i]].append(lineVals[curIdx])
        curIdx += 1

file.close()
```

```python
outputPath = '../Data/Raw/parsed_books_data.tsv'

with open(outputPath, 'w', encoding='utf-8') as f:
    headers = list(cols.keys())
    
    f.write('\t'.join(headers) + '\n')

    num_rows = len(cols[headers[0]])
    
    for i in range(num_rows):
        row_values = []
        for header in headers:
            value = cols[header][i]

            if header == 'genres':
                value = "|".join(value)
            else:
                value = str(value)
                value = value.replace('\t', ' ').replace('\n', ' ')

            row_values.append(value)

        f.write('\t'.join(row_values) + '\n')

print(f"Successfully saved {num_rows} rows to {outputPath}")
```
